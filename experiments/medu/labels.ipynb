```python
import pandas as pd
```

```python
engineering_sample_path = "gs://marin-us-east5/documents/medu-labels/mmlu-engineering/000000_000000.jsonl.gz"
sciences_sample_path = "gs://marin-us-east5/documents/medu-labels/mmlu-science/000000_000000.jsonl.gz"
social_sciences_sample_path = "gs://marin-us-east5/documents/medu-labels/mmlu-social-sciences/000000_000000.jsonl.gz"
mathematics_sample_path = "gs://marin-us-east5/documents/medu-labels/mmlu-mathematics/000000_000000.jsonl.gz"
# humanities_sample_path = "gs://marin-us-east5/documents/medu-labels/mmlu-humanities/000000_000000.jsonl.gz"
```

```python
score_options_dict = {
    "Useless": 1,
    "Poor": 2,
    "Okay": 3,
    "Good": 4,
    "Great": 5
}

def extract_score(row):
    text = row['generated_text']
    if "Final Score: " in text:
        score_text = text.split("Final Score: ")[1].split("\n")[0]
        for key, value in score_options_dict.items():
            if key in score_text:
                return value
    else:
        return None

def print_basic_stats(file_path):
    print(f"Printing basic stats for {file_path}")
    df = pd.read_json(file_path, lines=True, compression="gzip")
    df['score'] = df.apply(lambda row: extract_score(row), axis=1)
    print("Number of samples that could not be extracted: ", len(df[df['score'].isna()]))
    df = df[df['score'].notna()]
    score_distribution = df['score'].value_counts()
    print(score_distribution)
    great_df = df[df['score'] == 5]
    print(great_df['text'].iloc[0])
    return df
```

```python
filepaths = [
    engineering_sample_path,
    sciences_sample_path,
    social_sciences_sample_path,
    mathematics_sample_path,
    # humanities_sample_path,
]
for filepath in filepaths:
    print_basic_stats(filepath)
```
